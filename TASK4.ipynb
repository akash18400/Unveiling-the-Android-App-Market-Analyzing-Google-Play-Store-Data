## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

You can import your own data into Colab notebooks from your Google Drive account, including from spreadsheets, as well as from Github and many other sources. To learn more about importing data, and how Colab can be used for data science, see the links below under [Working with Data](#working-with-data).

````
import pandas as pdimport
import matplotlib.pyplot as plt
import plotly.express as px

# Load the datasets
apps_df = pd.read_csv('apps.csv')
reviews_df = pd.read_csv('user_reviews.csv')

# ----------------------------------------
# DATA CLEANING
# ----------------------------------------

# Drop unnamed index column if present
if 'Unnamed: 0' in apps_df.columns:
    apps_df.drop(columns=['Unnamed: 0'], inplace=True)

# Remove rows with missing key info
apps_df.dropna(subset=['Rating', 'Reviews', 'Size', 'Installs', 'Price'], inplace=True)

# Clean 'Installs' column
apps_df['Installs'] = apps_df['Installs'].str.replace('[+,]', '', regex=True).astype(int)

# Clean 'Price' column
apps_df['Price'] = apps_df['Price'].astype(str).str.replace('$', '', regex=False).str.strip()
apps_df['Price'] = pd.to_numeric(apps_df['Price'], errors='coerce')  # convert to float
apps_df.dropna(subset=['Price'], inplace=True)

# Convert 'Reviews' to int
apps_df['Reviews'] = apps_df['Reviews'].astype(int)

# Convert 'Size' from string to numeric (in KB)
def convert_size(size):
    if isinstance(size, str):
        if 'M' in size:
            return float(size.replace('M', '').strip()) * 1024
        elif 'k' in size:
            return float(size.replace('k', '').strip())
        elif 'Varies with device' in size:
            return None
    return pd.to_numeric(size, errors='coerce')

apps_df['Size'] = apps_df['Size'].apply(convert_size)
apps_df.dropna(subset=['Size'], inplace=True)
apps_df.reset_index(drop=True, inplace=True)

# ----------------------------------------
# CATEGORY EXPLORATION
# ----------------------------------------

category_counts = apps_df['Category'].value_counts()

plt.figure(figsize=(12, 6))
category_counts.plot(kind='bar')
plt.title('Number of Apps per Category')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# ----------------------------------------
# METRICS ANALYSIS
# ----------------------------------------

# 1. Average Rating per Category
avg_rating = apps_df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
avg_rating.plot(kind='bar', color='skyblue')
plt.title('Average App Rating per Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.ylim(3.5, 5)
plt.tight_layout()
plt.show()

# 2. Price vs Rating
plt.figure(figsize=(10, 6))
plt.scatter(apps_df['Price'], apps_df['Rating'], alpha=0.5)
plt.title('Price vs Rating')
plt.xlabel('Price ($)')
plt.ylabel('Rating')
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. Size vs Installs
plt.figure(figsize=(10, 6))
plt.scatter(apps_df['Size'], apps_df['Installs'], alpha=0.5)
plt.title('App Size vs Number of Installs')
plt.xlabel('Size (KB)')
plt.ylabel('Installs')
plt.grid(True)
plt.tight_layout()
plt.show()

# ----------------------------------------
# SENTIMENT ANALYSIS
# ----------------------------------------

# Drop missing sentiments
reviews_df.dropna(subset=['Sentiment', 'Sentiment_Polarity'], inplace=True)

# Average sentiment polarity per app
sentiment_summary = reviews_df.groupby('App')['Sentiment_Polarity'].mean().reset_index()
sentiment_summary.columns = ['App', 'Average_Sentiment_Polarity']

# Merge with main dataset
merged_df = pd.merge(apps_df, sentiment_summary, on='App', how='left')

# Plot: Rating vs Sentiment
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['Rating'], merged_df['Average_Sentiment_Polarity'], alpha=0.5)
plt.title('App Rating vs Average Sentiment Polarity')
plt.xlabel('Rating')
plt.ylabel('Sentiment Polarity')
plt.grid(True)
plt.tight_layout()
plt.show()

# ----------------------------------------
# INTERACTIVE VISUALIZATION (PLOTLY)
# ----------------------------------------

# 1. Interactive Price vs Rating
fig1 = px.scatter(apps_df, x='Price', y='Rating', hover_name='App', title='Price vs Rating (Interactive)')
fig1.show()

# 2. Interactive Installs vs Size
fig2 = px.scatter(apps_df, x='Size', y='Installs', hover_name='App', title='Size vs Installs (Interactive)', log_y=True)
fig2.show()

# 3. Interactive Sentiment Treemap
sentiment_heatmap = reviews_df.groupby(['App', 'Sentiment']).size().reset_index(name='Counts')
fig3 = px.treemap(sentiment_heatmap, path=['Sentiment', 'App'], values='Counts', title='Sentiment Distribution per App')
fig3.show()
````
